### Can we visualize where each athlete was born on a map?

```python
bios.head()
```

```python
len(bios)
```

```python
from geopy.geocoders import Nominatim
import time
cache = {}

def get_coordinates(city, region, country):
    geolocator = Nominatim(user_agent="my_app")
    key = f"{city}, {region}"
    print(f"Processing {city}, {region}")
    if key in cache:
        return cache[key]

    location = geolocator.geocode(key)

    if location:
        latitude = location.latitude
        longitude = location.longitude
        cache[key] = (latitude,longitude)
        return latitude, longitude
    else:
        print("No coordinates found")
        return None

bios_locs = bios.copy()

bios_locs['lat'] = None
bios_locs['long'] = None

for index, row in bios.iterrows():
    coordinates = get_coordinates(row['born_city'],row['born_region'], row['born_country'])
    if coordinates:
        lat, long = coordinates
        print(lat,long)
        bios_locs.loc[index,'lat'] = lat
        bios_locs.loc[index,'long'] = long

    if index % 10000 == 0:
        bios_locs.to_csv(f'updated_bios_{index}.csv', index=False)
        break
```

```python
# Example usage
city = "Madrid"
region = "Espana"
country = ""
coordinates = get_coordinates(city, region, country)
if coordinates:
    latitude, longitude = coordinates
    print(f"The coordinates of {city}, {region}, {country} are: Latitude: {latitude}, Longitude: {longitude}")
else:
    print("Coordinates not found for the specified location.")
```
